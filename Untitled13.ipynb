```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
import numpy as np # Needed for histogram edgecolors

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Create a Pandas DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['species'] = y
df['species'] = df['species'].apply(lambda x: target_names[x])

# Set a style for the plots
sns.set_style("whitegrid")
```

```python
# --- 1. Scatter Plot  ---
plt.figure(figsize=(8, 5))
colors = ['navy', 'turquoise', 'darkorange']
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(df.loc[df['species'] == target_name, 'sepal length (cm)'],
                df.loc[df['species'] == target_name, 'sepal width (cm)'],
                color=color, alpha=.8, lw=2, label=target_name)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('1. Scatter Plot: Sepal Length vs. Sepal Width by Species')
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.show()
```

```python
# --- 2. Histogram ---
plt.figure(figsize=(8, 5))
plt.hist(df['petal length (cm)'], bins=20, edgecolor='black')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('2. Histogram: Distribution of Petal Length')
plt.show()
```

```python
# --- 3. Box Plot ---
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal width (cm)', data=df)
plt.ylabel('Petal Width (cm)')
plt.title('3. Box Plot: Petal Width by Species')
plt.show()
```

```python
# --- 4. Violin Plot ---
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal length (cm)', data=df)
plt.title('4. Violin Plot: Sepal Length by Species')
plt.show()
```

```python
# --- 5. Heatmap of Correlation Matrix ---
plt.figure(figsize=(7, 5))
correlation_matrix = df.drop('species', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('5. Heatmap: Correlation Matrix of Iris Features')
plt.show()
```

```python
# --- 6. Pair Plot (using Seaborn) ---
# Note: Pair plots can take some time to generate for larger datasets
sns.pairplot(df, hue='species', diag_kind='kde') # diag_kind='kde' shows kernel density estimate on the diagonal
plt.suptitle('6. Pair Plot: Relationships Between Iris Features by Species', y=1.02)
plt.show()
```

```python
# --- 7. Distplot (Distribution Plot - using Seaborn) ---
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='sepal width (cm)', kde=True) # histplot is the newer function
plt.title('7. Distribution Plot: Sepal Width with KDE')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')
plt.show()
```

```python
# --- 8. Joint Plot (using Seaborn) ---
sns.jointplot(x='petal length (cm)', y='petal width (cm)', data=df, kind='reg') # kind='reg' adds regression line and marginal distributions
plt.suptitle('8. Joint Plot: Petal Length vs. Petal Width', y=1.02)
plt.show()
```

```python
# --- 9. Swarm Plot (using Seaborn) ---
plt.figure(figsize=(8, 6))
sns.swarmplot(x='species', y='petal length (cm)', data=df)
plt.title('9. Swarm Plot: Petal Length by Species')
plt.show()
```

```python
# --- 10. Count Plot (using Seaborn) ---
plt.figure(figsize=(7, 5))
sns.countplot(x='species', data=df)
plt.title('10. Count Plot: Number of Samples per Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
